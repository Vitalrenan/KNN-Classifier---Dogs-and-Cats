```python
#Joining the cats and dogs data frames
df3 = pd.concat([df2,df], axis=0)
df3
```

```python
#Plotting the data frame
sns.pairplot(df3,hue='dog',palette='coolwarm')
```

Conclusion: This figure shows that is possible to apply a KNN model over the dataset because almost all the features have different pattern between cats and dogs

```python
#Importing the Scaler lib from SkLearn
from sklearn.preprocessing import StandardScaler

#Scaling the features based on the main atribute Cats or Dogs
scaler = StandardScaler()
scaler.fit(df3.drop('dog',axis=1))
scaled_features = scaler.transform(df3.drop('dog',axis=1))

#updating the data frame values
df3_feat = pd.DataFrame(scaled_features,columns=df3.columns[:-1])
df3_feat.head()
```

```python
#Importing the test separation module from Sklean and applying 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_features,df3['dog'], test_size=0.30)
```

```python
#Importing the KNN module from SkLearn and applying
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
#Results of the model
pred = knn.predict(X_test)
pred

#Creating a data frame to compare the results
Cols = "Predictions,Correct Result,Result".split(',')

dfp = pd.DataFrame(index=list(range(1,len(pred)+1)),columns=Cols)
dfp["Predictions"] = pred
dfp["Correct Result"] = list(y_test)
dfp.loc[dfp["Predictions"] == dfp["Correct Result"], "Result"] = "Correct"
dfp.loc[dfp["Predictions"] != dfp["Correct Result"], "Result"] = "Incorrect"
dfp
```

```python
dfp['Result'].value_counts()
```

Conclusion: The model got 100% of the results right

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
```

The confusion matrix reveals that were 58 dogs and 16 cats on the test batch, and all of them were correctly predicted.

```python
print(classification_report(y_test,pred))
```

Therefore the precision and f1-score were 100%
